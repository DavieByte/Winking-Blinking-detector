```python
# adapted from https://github.com/python-pillow/Pillow/issues/3011#issuecomment-552141447
def convert_I_to_L(img):
    arrayd = np.uint8(np.array(img) / 256)
    return Image.fromarray(arrayd).convert("L")
```
